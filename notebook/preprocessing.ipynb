```python
import yfinance as yf
import pandas as pd
```

Gather historical data for TSLA, BND, and SPY from YFinance

```python
assets = ['TSLA', 'BND', 'SPY']
# Download historical data
data = yf.download(assets, start="2015-01-01", end="2024-10-31")
```

```python
data.head()
```

```python
# Access the closing prices of TSLA
tsla_close = data['Close']['TSLA']

# Access the volume of BND
bnd_volume = data['Volume']['BND']
```

 Clean and Understand the Data

```python
# Check for missing values
missing_values = data.isnull().sum()
print("Missing values per column:")
missing_values
```

```python
# Forward fill missing values (propagate the last valid value forward)
data.fillna(method='ffill', inplace=True)
```

```python
import pandas as pd

# Load the data with the first three rows as the header
df = pd.read_csv(r'C:\Users\Yibabe\Desktop\10academy_kifiya-AIMweek-11\data\yfinance_data.csv', header=[0, 1, 2])
```

Rename Columns More Readably:

```python
# Renaming columns for better clarity
df1.rename(columns={
    'Price Ticker Date': 'Date',
    'Adj Close BND Unnamed: 1_level_2': 'Adj Close BND',
    'Adj Close SPY Unnamed: 2_level_2': 'Adj Close SPY',
    'Adj Close TSLA Unnamed: 3_level_2': 'Adj Close TSLA',
    'Close BND Unnamed: 4_level_2': 'Close BND',
    'Close SPY Unnamed: 5_level_2': 'Close SPY',
    'Close TSLA Unnamed: 6_level_2': 'Close TSLA',
    'High BND Unnamed: 7_level_2': 'High BND',
    'High SPY Unnamed: 8_level_2': 'High SPY',
    'High TSLA Unnamed: 9_level_2': 'High TSLA',
    'Low BND Unnamed: 10_level_2': 'Low BND',
    'Low SPY Unnamed: 11_level_2': 'Low SPY',
    'Low TSLA Unnamed: 12_level_2': 'Low TSLA',
    'Open BND Unnamed: 13_level_2': 'Open BND',
    'Open SPY Unnamed: 14_level_2': 'Open SPY',
    'Open TSLA Unnamed: 15_level_2': 'Open TSLA',
    'Volume BND Unnamed: 16_level_2': 'Volume BND',
    'Volume SPY Unnamed: 17_level_2': 'Volume SPY',
    'Volume TSLA Unnamed: 18_level_2': 'Volume TSLA'
}, inplace=True)
```

```python
# Check the cleaned column names
df1.columns
```

Visualize the Data

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.plot(df1['Date'],df1['Close TSLA'], label='TSLA', color='blue')
plt.plot(df1['Date'], df1['Close BND'], label='BND', color='green')
plt.plot(df1['Date'], df1['Close SPY'], label='SPY', color='red')

plt.title('Closing Prices of TSLA, BND, and SPY')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()
```

```python
# Plot histograms of closing prices for TSLA, BND, and SPY
plt.figure(figsize=(15, 6))

plt.subplot(1, 3, 1)
df1['Close TSLA'].hist(bins=50, color='blue', alpha=0.7)
plt.title('TSLA Closing Price Distribution')
plt.xlabel('Price')
plt.ylabel('Frequency')

plt.subplot(1, 3, 2)
df1['Close BND'].hist(bins=50, color='green', alpha=0.7)
plt.title('BND Closing Price Distribution')
plt.xlabel('Price')
plt.ylabel('Frequency')

plt.subplot(1, 3, 3)
df1['Close SPY'].hist(bins=50, color='red', alpha=0.7)
plt.title('SPY Closing Price Distribution')
plt.xlabel('Price')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```
